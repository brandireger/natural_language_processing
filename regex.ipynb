### Write a function named is_vowel. It should accept a string as input and use a regular expression to determine if the passed string is a vowel. While not explicity mentioned in the lesson, you can treat the result of re.search as a boolean value that indicates whether or not the regular expression matches the given string.

```python
def is_vowel(string):
    found = re.search(r'[aeiouAEIOU]', string)
    if found:
        print(string, 'contains a vowel')
    else:
        print(string, 'does not contain a vowel')
```

```python
is_vowel('ant')
```

### Write a function named is_valid_username that accepts a string as input. A valid username starts with a lowercase letter, and only consists of lowercase letters, numbers, or the _ character. It should also be no longer than 32 characters. The function should return either True or False depending on whether the passed string is a valid username.

```python
regexp = r'(^[a-z])([^A-Z]){,31}$'
re.search(regexp, 'aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa')
```

```python
def is_valid_username(string):
    regexp = r'(^[a-z])([^A-Z]){,31}$'
    valid = re.search(regexp, string)
    if valid:
        return True
    else:
        return False
```

```python
assert is_valid_username('aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa') == False
assert is_valid_username('codeup') == True
assert is_valid_username('Codeup') == False
assert is_valid_username('codeup123') == True
assert is_valid_username('1codeup') == False
```

### Write a regular expression to capture phone numbers. It should match all of the following:


- (210) 867 5309
- +1 210.867.5309
- 867-5309
- 210-867-5309

```python
regexp = r'''
(?P<country_code>\d)?
\D*?
(?P<area_code>\d{3})?
\D*?
(?P<prefix>\d{3})
\D*?
(?P<line>\d{4})$
'''

match = re.search(regexp, '867-5309', re.VERBOSE)
match.groupdict()
```

```python
phones = ['(210) 867 5309', '+1 210.867.5309', '867-5309', '210-867-5309']

df = pd.DataFrame()
df['original_phone'] = pd.Series(phones)
df
```

```python
df.original_phone.str.extract(regexp, re.VERBOSE)
```

### Use regular expressions to convert the dates below to the standardized year-month-day format.


- 02/04/19
- 02/05/19
- 02/06/19
- 02/07/19
- 02/08/19
- 02/09/19

Use this file to answer the following questions:

- How many words have at least 3 vowels?

```python
regexp = r'([aeiou])'
re.findall(regexp, 'aardvark', re.IGNORECASE)
```

```python
def has_more_than_two_vowels(list):
    score = 0
    regexp = r'([aeiou])'
    for word in list:
        valid = re.findall(regexp, word, re.IGNORECASE)
        if len(valid) >= 3:
            score += 1
    return score
```

```python
has_more_than_two_vowels(wordlist)
```

- How many words have at least 3 vowels in a row?

```python
regexp = r'([aeiou]{3,})'
re.search(regexp, 'Abietineae', re.IGNORECASE)
```

```python
def has_three_vowels_in_row(list):
    score = 0
    regexp = r'([aeiou]{3,})'
    for word in list:
        valid = re.search(regexp, word, re.IGNORECASE)
        if valid:
            score += 1
    return score
```

```python
has_three_vowels_in_row(wordlist)
```

- How many words have at least 4 consonants in a row?

```python
def has_four_consonants_in_row(list):
    score = 0
    regexp = r'([^aeiou]{4,})'
    for word in list:
        valid = re.search(regexp, word, re.IGNORECASE)
        if valid:
            score += 1
    return score
```

```python
has_four_consonants_in_row(wordlist)
```
